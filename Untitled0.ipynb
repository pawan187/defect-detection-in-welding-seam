```python
import cv2
import numpy as np
```

```python
net = cv2.dnn.readNet("/content/drive/MyDrive/yolo_custom_model_Training/backup/yolov3_custom_7000.weights", "/content/drive/MyDrive/yolo_custom_model_Training/darknet/cfg/yolov3_custom.cfg")
classes = []
with open("/content/drive/MyDrive/yolo_custom_model_Training/custome_data/classes.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes)+1, 3))
```

```python
img = cv2.imread("/content/drive/MyDrive/IMG_20180515_155856.jpg")
img = cv2.resize(img, None, fx=0.4, fy=0.4)
height, width, channels = img.shape
```

```python
blob = cv2.dnn.blobFromImage(img, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
net.setInput(blob)
outs = net.forward(output_layers)
outs
```

```python
# Showing informations on the screen
class_ids = []
confidences = []
boxes = []
for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]
        if confidence > 0.2:
            # Object detected
            center_x = int(detection[0] * width)
            center_y = int(detection[1] * height)
            w = int(detection[2] * width)
            h = int(detection[3] * height)
            # Rectangle coordinates
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)
            boxes.append([x, y, w, h])
            confidences.append(float(confidence))
            class_ids.append(class_id)
boxes
```

```python
indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.1, 0.4)
indexes
```

```python
from google.colab.patches import cv2_imshow
font = cv2.FONT_HERSHEY_PLAIN
for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        color = colors[0]
        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
        cv2.putText(img,str(h)+'*'+str(w) ,(x, y +30),font,2,color)
cv2_imshow(img)
```
